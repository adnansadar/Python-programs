```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3],[5,7,4],"ro")
plt.plot([1.5,2.5,3.5],[5,7,4],"bx")
plt.show()
```

```python
x=np.arange(1,11)
y=2*(x+5)
plt.title("Matplot Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,y,".")
plt.show()
```

```python
x=np.arange(1,11)
y=np.sin(x)
plt.title("Matplot Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,y,"*")
plt.show()
```

```python
x=np.arange(1,11,0.01)
y=np.sin(x)
plt.title("Matplot Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.01)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin,"--k")#for dotted line
plt.title("Sine Waveform")

plt.subplot(2,2,2)
plt.plot(x,y_cos,"--k")
plt.title("Cos Waveform")

plt.show()
```

```python
x=[1,2,3]
y=[5,7,4]

x2=[1,2,3]
y2=[10,14,12]
plt.title("PROFIT AND LOSS",fontweight="bold",color="b")
plt.plot(x,y,label="Loss per month",color="r")
plt.plot(x2,y2,label="Profit per month",color="g")

plt.xlabel("Months",fontsize="xx-large",color='c')
plt.ylabel("Lakhs",fontsize="xx-large",color='m')
plt.legend()
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="Loss pm",width=0.5)

x2=[2,4,6,8,10]
y2=[8,6,2,5,6]
plt.title("PROFIT AND LOSS",fontweight="bold")
plt.bar(x2,y2,label="Profit pm",color="y",width=0.5)

plt.xlabel("Months",fontsize="xx-large",color='r')
plt.ylabel("Lakhs",fontsize="xx-large",color='b')
plt.legend()
plt.show()
```

```python
plt.barh([1,3,5,7,9],[5,2,7,8,2],label="Loss pm",height=0.5)
plt.barh([2,4,6,8,10],[8,6,2,5,6],label="Profit pm",color='y',height=0.5)
plt.title("PROFIT AND LOSS",fontweight="bold")

plt.xlabel("Months",fontsize="xx-large",color='r')
plt.ylabel("Lakhs",fontsize="xx-large",color='b')
plt.legend()
plt.show()
```

```python
slices=[7,2,3,12]
activities=['Sleeping','Eating','Coding','Watching']
cols=['c','m','g','y']

plt.pie(slices,
       labels=activities,
       wedgeprops={'linewidth': 3},
        colors=cols,
        startangle=90,
        shadow='True',
        explode=(0,0,0.2,0))#used to slice a pie out of the chart
plt.title("Repeat Mode",fontweight='bold')
plt.legend()
plt.show()
```

```python
slices=[7,2,3,12]
activities=['Sleeping','Eating','Coding','Watching']
cols=['c','m','g','y']

plt.pie(slices,
       labels=activities,
       wedgeprops={'linewidth': 3},
       colors=cols,
       startangle=90,
       shadow='True',
       explode=(0,0,0.2,0),#used to slice a pie out of the chart
       autopct='%1.1f%%')
plt.title("Repeat Mode",fontweight='bold')
plt.legend()
plt.show()
```
